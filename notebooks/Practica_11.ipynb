## Operador IN

### Crea un subquery para generar una vista con:

**1 Los vendedores que no vendieron en 2018**

```sql
SELECT *
FROM base.Vendedor AS v
WHERE id_vendedor NOT IN (
    SELECT DISTINCT id_vendedor
    FROM base.ventas AS ve
    WHERE YEAR(str_to_date(ve.fecha_venta,'%d-%b-%y')) = 2018
);


```

**2 Productos que no se vendieron en marzo del 2017**

```sql
SELECT p.id_producto, p.producto
FROM base.producto AS p
WHERE id_producto NOT IN (
    SELECT DISTINCT id_producto
    FROM base.ventas AS v 
    WHERE (MONTH(str_to_date(v.Fecha_venta,'%d-%b-%y')) = 3) AND (YEAR(str_to_date(v.Fecha_venta,'%d-%m-%y')) = 2017)
);

  
```

**3. Clientes que compraron propiedas de remates Bancarios en el primer cuarto de 2016**

```sql
SELECT v.id_Cliente, v.nombre
FROM base.clientes AS v
WHERE v.id_cliente IN (
    
    SELECT DISTINCT id_cliente
    FROM base.ventas AS ve 
    WHERE (YEAR(STR_TO_DATE(ve.Fecha_venta,'%d-%b-%y'))=2016) AND (ve.id_producto=7) AND (MONTH(STR_TO_DATE(ve.Fecha_venta,'%d-%b-%y'))<4)
);
```

### <span style="font-size: 16.38px;">Uso de ALL &amp; ANY</span>

### <span style="font-size: 16.38px;">Genera una vista para entregar el informe que contenga:</span>

**4. Todas las ventas superiores a la venta más cara de una oficina**

```sql
SELECT *
FROM base.ventas AS v
 WHERE v.Venta > ALL(
    SELECT Venta FROM base.ventas AS ve
    WHERE ve.ID_producto = 2
 );
```

**5. Cualquier venta mayor al de una oficina**

```sql
SELECT *
FROM base.ventas AS v
 WHERE v.Venta > ANY(
    SELECT Venta FROM base.ventas AS ve
    WHERE ve.ID_producto = 2
 );
```

## EXISTS

## Buscaremos a todos los vendedores que no vendieron usando EXISTS y el uso de un query correlacionado:

**6. Todos los vendedores que no han vendido nada**

```sql
SELECT Nombre, Apellido
FROM base.vendedor AS ve
 WHERE NOT EXISTS(
    SELECT v.ID_Vendedor
    FROM base.ventas v
    WHERE ve.ID_Vendedor = v.ID_Vendedor); 
```

## Subqueries en SELECT

## Crea un subquery en SELECT para:

**7. Obtener el % que representa cada venta del total por cliente**

```sql
SELECT 
    v.id_venta,
    v.id_cliente,
    v.venta,
    (v.venta / va.total_venta_cliente) * 100 AS porcentaje_venta_cliente
FROM 
    base.ventas v
JOIN (
    SELECT 
        id_cliente,
        SUM(venta) AS total_venta_cliente
    FROM 
        ventas
    GROUP BY 
        id_cliente) as va
    ON  v.id_cliente = va.id_cliente;
```
